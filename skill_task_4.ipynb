```python
# Basic Key Logging Program (Educational Purpose)

print("Basic Key Logger")
print("Type something and press ENTER")
print("Type 'exit' to stop\n")

# Open a file in append mode
log_file = open("key_log.txt", "a")

while True:
    user_input = input("Enter text: ")

    if user_input.lower() == "exit":
        print("Key logging stopped.")
        break

    # Log each character separately
    for key in user_input:
        log_file.write(key)

    # Add new line after each input
    log_file.write("\n")

log_file.close()
```
